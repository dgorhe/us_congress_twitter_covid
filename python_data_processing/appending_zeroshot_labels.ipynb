# COVID Labels

```python
ls = []

for filename in os.listdir(covid_labeled_json):
    f = os.path.join(covid_labeled_json, filename)
    
    if os.path.isfile(f):
        ls.append(f)
        
main = []

for file in ls:
    tmp = json.load(open(file, "r"))
    main.append(tmp)
    
    
main2 = [y for x in main for y in x ]
labels = []


for tweet in main2:
    if tweet["scores"][0] > 0.5:
        labels.append("covid")
    else:
        labels.append("non-covid")
```

# Location Labels

```python
location_labeled_sample_json = "/Users/darvesh/Programming/R/fall2021/gr5702/us_congress_twitter_covid/data/created/sample/labeled_json/location_labeled_tweets_first_1000"

ls = []

for filename in os.listdir(location_labeled_sample_json):
    f = os.path.join(location_labeled_sample_json, filename)
    
    if os.path.isfile(f):
        ls.append(f)
        
ls.sort()

main = []

for file in ls:
    tmp = json.load(open(file, "r"))
    main.append(tmp)
    
main2 = [y for x in main for y in x ]

location_labels = []

# for tweet in main2:
#     if (tweet["scores"][0] > 0.5):
#         location_labels.append((tweet["sequence"], tweet["labels"][0]))
#     else:
#         location_labels.append((tweet["sequence"], "uncategorized"))

for tweet in main2:
    if (tweet["scores"][0] > 0.5):
        location_labels.append(tweet["labels"][0])
    else:
        location_labels.append("uncategorized")

```
